```python
ds=pd.read_csv("spam.csv" ,encoding='latin-1')
pd.set_option("display.max_colwidth",100)
ds
```

```python
ds=ds.drop(labels=["Unnamed: 3","Unnamed: 4", "Unnamed: 2"],axis=1)
```

```python
ds.columns=["lables","Text"]
```

                                                                    Removing punctuation

```python
import string
string.punctuation
```

```python
def remove_pun(text):
    text="".join([char for char in text if char not in string.punctuation])
    return text 
ds["text_clean"]=ds["Text"].apply(lambda X:remove_pun(X))
```

```python
ds.head()
```

                                                                        Tokenizing

```python
import re

def token(text):
    tokens=re.split("\W+",text)
    return tokens
ds["text_token"]=ds["text_clean"].apply(lambda x:token(x.lower()))
```

```python
ds.head()
```

                                                                Removing stop words

```python
stopwords=nltk.corpus.stopwords.words("english")
def remove_stopwords(text_tkt):
    text=[word for word in text_tkt if word not in stopwords]
    return text
ds["no_stopwords"]=ds["text_token"].apply(lambda x:remove_stopwords(x))
```

```python
ds.head()
```

                                                                    TF-IDF vectorization


```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf=TfidfVectorizer(analyzer=remove_stopwords)
x_tfidf=tfidf.fit_transform(ds["no_stopwords"])
print(x_tfidf.shape)
tfidf.get_feature_names()
```

```python
x_tfidf
```

```python
X_features=pd.DataFrame(x_tfidf.toarray())
X_features.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_score,recall_score
```

```python
x_train,x_test,y_train,y_test=train_test_split(X_features,ds["lables"],test_size=0.2)
```

```python
rf_model=rf.fit(x_train,y_train)
rf_predict=rf_model.predict(x_test)
```

```python
rf_predict
```

```python
precision = precision_score(y_test, rf_predict, pos_label='spam')
recall = recall_score(y_test, rf_predict, pos_label='spam')
```
